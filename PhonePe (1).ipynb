```python
# Calculate the total number of transactions and total transaction amount for each state over the years.
state_txn_and_users
```

```python
print('the total number of transactions and total transaction amount for each state over the years:')
total_txns=state_txn_and_users.groupby('State').agg({
    'Transactions': 'sum',
    'Amount (INR)': 'sum'
}).reset_index()
total_txns
```

```python
# Identify the top 5 states with the highest transaction volumes and the top 5 states with the lowest transaction volumes.
txns_volume=total_txns[['State','Amount (INR)']].sort_values(by='Amount (INR)',ascending=False)
txns_volume
```

```python
print("Top 5 States with the highest transaction volumes:")
txns_volume.head()
```

```python
print("Top 5 States with the lowest transaction volumes:")
txns_volume.tail()
```
