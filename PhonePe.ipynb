```python
#Merge the State_Txn and Users dataset with the District Demographics dataset
#to calculate the ratio of registered users to the population for each state
state_registered_users=state_txn_and_users.groupby('State')['Registered Users'].sum().reset_index()
state_registered_users
```

```python
#getting population from district demographics
state_wise_population=district_demographics.groupby('State')['Population'].sum().reset_index()
state_wise_population
```

```python
#the ratio of registered users to the population for each state
users_to_population=state_registered_users.merge(state_wise_population, on ='State', how='left')
users_to_population
```

```python
# Create a column chart depicting the ratio of users to population by state
plt.figure(figsize=(10,6))
plt.barh(users_to_population['State'],users_to_population['Users to Population Ratio'])

plt.title('The ratio of users to population by state')
plt.xlabel('Users to Population Ratio')
plt.ylabel('State')
plt.show()
```

### Key Insights: Users to Population Ratio by State

- **Highest ratios in metro/UT regions** like Delhi (5.68), Dadra & Nagar Haveli (4.60), and Telangana (4.10), showing high app penetration relative to population.  
- **Moderate ratios** in large states like Andhra Pradesh (3.20), Maharashtra (2.81), and Gujarat (1.92), reflecting substantial but proportionally lower adoption.
