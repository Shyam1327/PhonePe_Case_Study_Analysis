```python
# Identify the district with the highest population for each state
district_demographics
```

```python
district_population=district_demographics[['State','District','Population']].copy()
district_population
```

```python
high_population_district=district_population.loc[district_population.groupby('State')['Population'].idxmax()]
high_population_district
```

```python
#sorting the districts bases on the most populated
high_population_district=high_population_district.sort_values(by='Population', ascending=False)
high_population_district
```

```python
# Create a column chart depicting the district with the highest population for each state
import matplotlib.pyplot as plt
plt.figure(figsize=(14,8))

label=[f"{d}-{s}" for d,s in zip(high_population_district['District'],high_population_district['State'])]
plt.barh(label,high_population_district['Population'])

plt.title('District with the highest population for each state', color='Orange')
plt.xlabel('Population')
plt.ylabel('State')
plt.show()
```
